```python
import re

texts = []
with open('speeches.txt', 'r', encoding='utf-8') as f:
    texts.extend(['SPEECH' + text for text in f.read().split('SPEECH')][1:])
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
tokenized_texts = []
for text in texts:
    tokenized_texts.append(word_tokenize(text.lower()))
```

```python
from collections import Counter
```

```python
from string import punctuation
noise = stopwords.words('english') + list(punctuation)
```

```python
counter = Counter()
for i, tokenized_text in enumerate(tokenized_texts):
    tokenized_texts[i] = [word for word in tokenized_text if word not in noise]
    counter.update(tokenized_text)
```

```python
# мы удалили stop-слова ранее, поэтому их мы не учитываем.
counter.most_common(10)
```

```python
# people и know -- ключевые слова. '’', '–' -- лишние знаки пунктуации, которых нет в string.punctuation и 'going' скорее тоже можно отнести к стоп-словам
counter.most_common(5)
```

```python
# функция для подсчёта tf слов в документе
def tfs(words):
    tf = Counter(words)
    return {word: tf[word] / len(words) for word in words}
```

```python
import math

# функция для подсчёта idf слова в документе
def idf(word, corpus):
    return math.log10(len(corpus) / (1 + len([1 for text in corpus if word in text]))) # added 1 so that we don't divide by zero
```

```python
# document 1
num_document = 0
document = tokenized_texts[num_document]
```

```python
term_frequencies = tfs(document)
for word in document:
    print('tf-idf of word ' + word + ': ' + str(term_frequencies[word] * idf(word, tokenized_texts)))
```

```python
sents = ['John built a house', 'Run away', 'John gave Mary roses', 'The curse has come upon me']
```

```python
def print_parses(parser, sentence):
    for elem in parser.parse(sentence.split()):
        print(elem)
```

Продемонстрируем алгоритм CYK parser-а

Матрица в CYK должна быть трёхмерной, но ради удобства сделаем её двумерной, схлопнув по оси Z.

```python
import pandas as pd
import numpy as np
n = 4
sent = sents[0].split()
df = pd.DataFrame(np.empty((n + 1, n)), columns=range(1, n + 1), dtype=str)
df = df.reindex(index=df.index[::-1])
df[:] = ''
```

На нулевую строку поместим слова нашего предложения

```python
df.iloc[n, :] = sent
df.style.set_properties(**{'text-align': 'center'})
```

Заполним первый ряд в соответствии с алгоритмом, т.е. пройдёмся с шагом 1 по первому ряду и запишем все возможные нетерминальные вершины, которые переходят в соответствующие терминальные

```python
df.iloc[n - 1, :] = np.array(('NP', 'V', 'DET', 'N'))
df.style.set_properties(**{'text-align': 'center'})
```

Заполним второй ряд в соответствии с алгоритмом, т.е. пройдёмся с шагом 2 по второму ряду и найдём все возможные разбиения на 2 составляющих, поскольку мы исходим из предположения о том, что грамматика в нормальной форме Хомского. В данном случае нашли только одно разложение: NP -> DET N

```python
df.iloc[n - 2, :] = np.array(('', '', 'NP', ''))
df.style.set_properties(**{'text-align': 'center'})
```

Заполним третий ряд в соответствии с алгоритмом, т.е. пройдёмся с шагом 3 по третьему ряду и найдём все возможные разбиения на 2 составляющих, поскольку мы исходим из предположения о то, что грамматика в нормальной форме Хомского. В данном случае нашли только одно разложение VP -> V NP

```python
df.iloc[n - 3, :] = np.array(('', 'VP', '', ''))
df.style.set_properties(**{'text-align': 'center'})
```

Наконец, проверим, что S правильно раскладывается и на последнем ряду. И действительно, S -> NP VP

```python
df.iloc[n - 4, :] = np.array(('S', '', '', ''))
df.style.set_properties(**{'text-align': 'center'})
```

Проведём анализ также для предложения "Flies fly quickly"

```python
n = 3
sent = 'Flies fly quickly'.split()
df = pd.DataFrame(np.empty((n + 1, n)), columns=range(1, n + 1), dtype=str)
df = df.reindex(index=df.index[::-1])
df[:] = ''
```

```python
# Добавим правила для этого предложения

rules = """
    S -> NP VP | V ADV
    VP -> V NP | V PREPP | V VP | V ADV
    PREPP -> PREP N
    NP -> DET N | NP N
    DET -> 'a' | 'The'
    ADV -> 'away' | 'quickly'
    PREP -> 'upon'
    N -> 'house' | 'roses' | 'curse' | 'me'
    NP -> 'John' | 'Mary' | 'Flies'
    V -> 'built' | 'Run' | 'gave' | 'has' | 'come' | 'fly'
""".split('\n')

grammar = nltk.CFG.fromstring('\n'.join(rules))
```

```python
df.iloc[n, :] = sent
df.style.set_properties(**{'text-align': 'center'})
```

```python
df.iloc[n - 1, :] = np.array(('NP', 'V', 'ADV'))
df.style.set_properties(**{'text-align': 'center'})
```

```python
df.iloc[n - 2, :] = np.array(('', 'VP', ''))
df.style.set_properties(**{'text-align': 'center'})
```

```python
# Как видно, это предложение порождается нашей грамматикой
df.iloc[n - 3, :] = np.array(('S', '', ''))
df.style.set_properties(**{'text-align': 'center'})
```

```python
sents.append(' '.join(sent))
cp = nltk.EarleyChartParser(grammar)
for sent in sents:
    print(str(sent) + ':')
    print_parses(cp, sent)
    print()
```
